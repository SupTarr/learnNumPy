# Matplotlib - Basic Chart Plotting

## Import necessary packages

```python
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt

# Jupyter magic command
%matplotlib inline
```

## 1. Basic Plotting

---

### 1.1 Prepare data

```python
x = np.arange(0, 20, 0.1)
sin = np.sin(x)
cos = np.cos(x)
```

### 1.2 Plot chart with `matplotlib.pyplot`

```python
# By default, pyplot.plot() draws a Line chart
plt.plot(x, sin)
plt.plot(x, cos)

plt.title('sin(x) and cos(x)')
plt.xlabel('x-value')
plt.ylabel('y-value')
```

### 2.2 Use Current Figure and Axes Explicitly

```python
# Get current figure
fig = plt.gcf()

# Get current axes
axes = plt.gca()

# plt.plot(x, sin)
# plt.plot(x, cos)

sin_plt = axes.plot(x, sin)
cos_plt = axes.plot(x, cos)

# plt.title('sin(x) and cos(x)')
axes.set_title('sin(x) and cos(x)', color = 'blue')

# plt.xlabel('x-value')
axes.set_xlabel('x-value', color = 'blue')

# plt.ylabel('y-value')
axes.set_ylabel('y-value', color = 'blue')
```

## 3. Working with subplots

---

### 3.1 Understand subplots layout

```python
# Default nrows = 1, ncols = 1
# Default figure size = (6.4, 4.8), dpi = 100

# fig, axes = plt.subplots()
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 6))

print(fig)
print(axes)
```

### 3.2 Working with Grid

```python
# Create figure
fig = plt.figure(figsize = (10, 8))

# Convert to grid
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan = 2)

# Draw chart on ax1
x1 = np.arange(0, 10, 0.5)
y1 = 3 * x1 + 5
ax1.plot(x1, y1)

# Draw chart on ax2
x2 = np.arange(0, 10, 0.5)
y2 = 8 * (x2 ** 2) - (2 * x2) -10
ax2.plot(x2, y2, color = 'orange')

# Draw chart on ax3
x3 = np.arange(-10, 10, 0.5)
y3 = -(x3 ** 3) + (x3 ** 2) + x3 -10
ax3.plot(x3, y3, color = 'blue')

rng = default_rng()
y4 = y3 + rng.integers(low = -200, high = 200, endpoint = True, size = 40)
ax3.scatter(x3, y4, color = 'brown')
```
